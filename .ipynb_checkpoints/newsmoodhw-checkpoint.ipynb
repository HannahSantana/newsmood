```python
media_sources = ['@BBC', '@CBS', '@FoxNews','@CNN','@nytimes']

merged_list = []

sentiments = []

counter = 1 

def media_sources_analysis(media_sources):
    
    global counter
    
    for media in media_sources:
        
        oldest_tweet = None
        
        media_tweets = tweepy.Cursor(api.user_timeline, id= media, max_id = oldest_tweet ).items(100)
        
        for tweet in media_tweets:
            
            tweet_json = tweet._json 
    
            results = analyzer.polarity_scores(tweet_json["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
            tweets_ago = counter

            oldest_tweet = tweet_json['id'] - 1
            
            sentiments.append({"Date": tweet_json["created_at"],
                               "Media": media, 
                               "Compound": compound,
                               "Positive": pos,
                               "Negative": neg,
                               "Neutral": neu,
                               "Tweets Ago": counter})
                                    
            counter += 1 
        
        merged_list.append(sentiments)
```

```python
media_sources_analysis(media_sources)
```
